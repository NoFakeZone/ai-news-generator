# Convert data from news.sqlite3 to JSON
This notebook retrieves all articles from the SQLite database and saves them in a format compatible with `scraped_news.json`.

```python
import sqlite3
import json

# 1. Connect to the database and set row_factory to sqlite3.Row to easily convert results to dictionaries
conn = sqlite3.connect('news.sqlite3')
conn.row_factory = sqlite3.Row
cursor = conn.cursor()

# 2. Fetch all data from the articles table
cursor.execute("SELECT * FROM articles")
rows = cursor.fetchall()

# 3. Convert each row to a dictionary
data = [dict(row) for row in rows]

# 4. Save the data to a new JSON file
with open('scraped_news_from_db.json', 'w', encoding='utf-8') as f:
    # ensure_ascii=False makes sure special characters are saved correctly
    # indent=4 nicely formats the file (similar to scraped_news.json)
    json.dump(data, f, ensure_ascii=False, indent=4)

print(f"Successfully saved {len(data)} articles to scraped_news_from_db.json")

# 5. Close the database connection
conn.close()
```
